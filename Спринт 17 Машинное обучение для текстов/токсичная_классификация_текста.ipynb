Сначала делаем: ```output = model(batch).logits```Логит - это необработанные результаты, полученные на последнем уровне нейронной сети до того, как произойдет активация.
    
    
Теперь просто берем максимум: ```_, predicted = torch.max(output, 1)``` - первое значение нам не нужно(это само максимальное число), а второе это как раз индекс, который соответствует нашему классу
    
    
Такой метод позволит сразу получить наши классы и обучать тут ничего не надо:(

Создание токенизатора и модели BERT.

````python
tokenizer = transformers.BertTokenizer.from_pretrained('unitary/toxic-bert')
model = transformers.BertModel.from_pretrained('unitary/toxic-bert')
````

Функция получения embeddings.

````python
def bert_embedding(column_text):
    embeddings = []
    tokenized = column_text.progress_apply(
        lambda x: tokenizer.encode(x, max_length=TEXT_SIZE, truncation=True, add_special_tokens=True))
    padded = pad_sequence([torch.as_tensor(seq) for seq in tokenized], batch_first=True)
    attention_mask = np.where(padded != 0, 1, 0)
    input_ids = torch.tensor(padded)
    attention_mask = torch.tensor(attention_mask.astype(np.int64))


    for i in tqdm(range(input_ids.shape[0] // BATCH_SIZE)):
        batch = torch.LongTensor(input_ids[BATCH_SIZE*i:BATCH_SIZE*(i+1)]).cuda() # закидываем тензор на GPU
        attention_mask_batch = torch.LongTensor(attention_mask[BATCH_SIZE*i:BATCH_SIZE*(i+1)]).cuda()

        with torch.no_grad():
            model.cuda()
            batch_embeddings = model(batch, attention_mask=attention_mask_batch)

        embeddings.append(batch_embeddings[0][:,0,:].cpu().numpy()) # перевод обратно на проц, чтобы в нумпай кинуть
        del batch
        del attention_mask_batch
        del batch_embeddings

    return np.concatenate(embeddings)
````

## Сравнение моделей на тестовой выборке

Сравним результаты предсказаний для лучших моделей на тестовой выборке и выберем модель с большим значением метрики F1.

### Функция анализа модели.
Функция производит обучение модели, предсказание на тестовой выборке, а также измеряет время процессов обучения и предсказания.

````python
def model_test(model):

    # обучение модели с замером времени 
    time_fit_start = time()
    model.fit(X_train, y_train)
    time_fit = time()-time_fit_start
    # предсказание модели с замером времени
    time_predict_start = time()
    predict = model.predict(X_test)
    time_predict = time()-time_predict_start
    
    return time_fit, time_predict, f1_score(y_test, predict), predict
````

### Применение функции к лучшим моделям.

````python
# шаблон таблицы для результатов
result_models_test = pd.DataFrame(columns=['model','time_fit','time_predict','f1_score'])
# шаблон таблицы для предсказаний
predict_models = pd.DataFrame(y_test)
comments_test_predict_models = comments_test

# применение функции к лучшим моделям
for best_model in [(best_model_LR,'Logistic Regression'), 
                   (best_model_RF,'Random Forest'), 
                   (best_model_CB,'CatBoost'), 
                   (best_model_LGBM,'LightGBM')]:
    time_fit,time_predict,score, predict = model_test(best_model[0])
    result_models_test = pd.concat([result_models_test,
                                    pd.DataFrame({'model':[best_model[1]],
                                                  'time_fit':[time_fit],
                                                  'time_predict':[time_predict],
                                                  'f1_score':[score]})
                                   ])
    comments_test_predict_models = pd.concat([comments_test_predict_models, 
                                pd.DataFrame(predict, columns=[f'predict_{best_model[1]}'], 
                                             index=y_test.index)], axis=1)
    
result_models_test.reset_index(drop=True, inplace=True)
````
