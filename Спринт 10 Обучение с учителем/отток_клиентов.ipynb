## –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –∑–∞–¥–∞—á–∏ (–ø–æ–ª—É—á–µ–Ω–∏–µ –ª—É—á—à–µ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞, –±–µ–∑ —É—á—ë—Ç–∞ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ –∫–ª–∞—Å—Å–æ–≤)

### –†–∞–∑–±–∏–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â—É—é, –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏

–í –ø—Ä–æ–µ–∫—Ç–µ –±—É–¥–µ—Ç –ø—Ä–∏–º–µ–Ω–µ–Ω–æ –æ–±—É—á–µ–Ω–∏–µ —Å —É—á–∏—Ç–µ–ª–µ–º–º, –¥–ª—è —ç—Ç–æ–≥–æ –≤—ã–¥–µ–ª–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∏ `features` –∏ —Ü–µ–ª–æ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ `target`.

```python
features = churn.drop(columns='Exited')
target = churn.Exited
```

–†–∞–∑–æ–±—å—ë–º –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –Ω–∞ 3 –≤—ã–±–æ–∫–∏ –≤ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–∏ 3:1:1 (–æ–±—É—á–∞—é—â—É—é, –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É –∏ —Ç–µ—Å—Ç–æ–≤—É—é —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ).

```python
# –≤—ã–¥–µ–ª–∏–º –æ–±—É—á–∞—é—â—É—é –≤—ã–±–æ—Ä–∫—É –≤–∫–ª—é—á–∞—é—â—É—é –≤ —Å–µ–±—è 60 % –¥–∞–Ω–Ω—ã—Ö
features_train, features_valid_test, target_train, target_valid_test = train_test_split(
    features, target, train_size=0.6, stratify=target, random_state=RANDOM)

# –æ—Å—Ç–∞–≤—à—É—é—Å—è —á–∞—Å—Ç—å —Ä–∞–∑–¥–µ–ª–∏–º –ø–æ–ø–æ–ª–∞–º –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
features_valid, features_test, target_valid, target_test = train_test_split(
    features_valid_test, target_valid_test, test_size=0.5, stratify=target_valid_test, random_state=RANDOM)
```

<div class='alert alert-success'> ‚úîÔ∏è–ö–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ! –†–µ–∫–æ–º–µ–Ω–¥—É—é, –µ—Å–ª–∏ –∑–Ω–∞–µ—à—å –æ –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ –¥–µ–ª–∏—Ç—å –¥–∞—Ç–∞—Å–µ—Ç, –∏—Å–ø–æ–ª—å–∑—É—è –ø–∞—Ä–∞–º–µ—Ç—Ä stratify, —á—Ç–æ–±—ã –ø–æ–ø—ã—Ç–∞—Ç—å—Å—è –æ–±–µ—Å–ø–µ—á–∏—Ç—å —Å–æ–±–ª—é–¥–µ–Ω–∏–µ –±–∞–ª–∞–Ω—Å–∞.(https://zen.yandex.ru/media/id/5ee6f73b7cadb75a66e4c7e3/sozdanie-vyborok-v-usloviiah-disbalansa-klassov-625654c41dd29a5c65819c90)  

</div>

<div class='alert alert-info'> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π, –¢–∞—à–∏—Ä–µ–≤–∞.
<br> –°–ø–∞—Å–∏–±–æ –∑–∞ –∑–∞–º–µ—á–∞–Ω–∏–µ. –ü—Ä–∏–º–µ–Ω–∏–ª –ø–∞—Ä–∞–º–µ—Ç—Ä stratify. –ò—Ç–æ–≥–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ F1-–º–µ—Ä—ã –≤–æ–∑—Ä–æ—Å–ª–æ —Å 0.60 –¥–æ 0.62.

</div>

```python
print('–†–∞–∑–º–µ—Ä –æ–±—É—á–∞—é—â–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:', features_train.shape)
print('–†–∞–∑–º–µ—Ä —Ü–µ–ª–µ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:', target_train.shape)
print('–†–∞–∑–º–µ—Ä –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:', features_valid.shape)
print('–†–∞–∑–º–µ—Ä —Ü–µ–ª–µ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤–∞–ª–∏–¥–∞—Ü–∏–∏:', target_valid.shape)
print('–†–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:', features_test.shape)
print('–†–∞–∑–º–µ—Ä —Ü–µ–ª–µ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Ç–µ—Å—Ç–∞:', target_test.shape)
```

### –ü–æ–ª—É—á–µ–Ω–∏–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–π –º–æ–¥–µ–ª–∏ –±–µ–∑ —É—á—ë—Ç–∞ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ –∫–ª–∞—Å—Å–æ–≤

#### –ú–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –¥–µ—Ä–µ–≤–æ–º —Ä–µ—à–µ–Ω–∏–π

–û–ø—Ä–µ–¥–µ–ª–∏–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞ `max_depth` –ø–µ—Ä–µ–±–æ—Ä–æ–º –≤ —Ü–∏–∫–ª–µ –∏ –æ—Ü–µ–Ω–∫–µ –ø–æ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –Ω–∞ F1-–º–µ—Ä–µ.

```python
best_model_tree = None
best_score_f1_tree = 0

for depth in tqdm(range(1, 11)):
    model = DecisionTreeClassifier(random_state=RANDOM, max_depth=depth)
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)
    score = f1_score(target_valid, predicted_valid)
    if score > best_score_f1_tree:
        best_model_tree = model
        best_score_f1_tree = score

print(f'–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞: {best_model_tree}\n'
      f'F1-–º–µ—Ä–∞: {best_score_f1_tree}\n'
      f'AUC-ROC: {roc_auc_score(target_valid, best_model_tree.predict_proba(features_valid)[:,1])}')
```

<div class='alert alert-success'> ‚úîÔ∏è–û—Ç–ª–∏—á–Ω–æ, –ø–æ—Å–º–æ—Ç—Ä–µ–ª–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–µ—Ä–µ–≤–∞ –∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä –ø—Ä–æ–≥—Ä–µ—Å—Å–∞.
</div>

#### –ú–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

–û–ø—Ä–µ–¥–µ–ª–∏–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞ `n_estimators` –ø–µ—Ä–µ–±–æ—Ä–æ–º –≤ —Ü–∏–∫–ª–µ –∏ –æ—Ü–µ–Ω–∫–µ –ø–æ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –Ω–∞ F1-–º–µ—Ä–µ, –ø–∞—Ä–∞–º–µ—Ç—Ä `max_depth` –ø—Ä–∏–º–µ–º —Ä–∞–≤–Ω—ã–º 7, –∫–∞–∫ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π –¥–ª—è —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞.

```python
best_model_forest = None
best_score_f1_forest = 0

for est in tqdm(range(1, 100, 2)):
    model = RandomForestClassifier(random_state=RANDOM, n_estimators=est, max_depth=7)
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)
    score = f1_score(target_valid, predicted_valid)
    if score > best_score_f1_forest:
        best_model_forest = model
        best_score_f1_forest = score

print(f'–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞: {best_model_forest}\n'
      f'F1-–º–µ—Ä–∞: {best_score_f1_forest}\n'
      f'AUC-ROC: {roc_auc_score(target_valid, best_model_forest.predict_proba(features_valid)[:,1])}')
```

#### –ú–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

–¢–∞–∫ –∫–∞–∫ –Ω–∞—à–∞ –æ–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞ –Ω–µ –≤–µ–ª–∏–∫–∞, –ø—Ä–∏–º–µ–Ω–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä `solver='liblinear'`.
–û–ø—Ä–µ–¥–µ–ª–∏–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ `penalty` –∏ `C` –ø–µ—Ä–µ–±–æ—Ä–æ–º –≤ —Ü–∏–∫–ª–µ –∏ –æ—Ü–µ–Ω–∫–µ –ø–æ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –Ω–∞ F1-–º–µ—Ä–µ.

```python
best_model_logistic = None
best_score_f1_logistic = 0

for penalty in ('l1', 'l2'):
    for inv_reg in tqdm(range(1, 10000, 20)):
        model = LogisticRegression(penalty=penalty, solver='liblinear', C=inv_reg/100, random_state=RANDOM)
        model.fit(features_train, target_train)
        predicted_valid = model.predict(features_valid)
        score = f1_score(target_valid, predicted_valid)
        if score > best_score_f1_logistic:
            best_model_logistic = model
            best_score_f1_logistic = score    

print(f'–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏: {best_model_logistic}\n'
      f'F1-–º–µ—Ä–∞: {best_score_f1_logistic}\n'
      f'AUC-ROC: {roc_auc_score(target_valid, best_model_logistic.predict_proba(features_valid)[:,1])}')
```

–ü—Ä–∏ –¥–∏—Å–±–∞–ª–∞–Ω—Å–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ F1-–º–µ—Ä–µ (0.56) –ø–æ–∫–∞–∑–∞–ª–∞ –º–æ–¥–µ–ª—å —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞.

AUC-ROC –º–µ—Ä–∞ –¥–ª—è –≤—Å–µ—Ö –º–æ–¥–µ–ª–µ–π –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –≤—ã—à–µ F1-–º–µ—Ä—ã.

### –ü—Ä–∏–º–µ—Ä –±–æ—Ä—å–±—ã —Å –¥–∏—Å–±–∞–ª–∞–Ω—Å–æ–º –∫–ª–∞—Å—Å–æ–≤, –ø—Ä–∏ –ø–æ–º–æ—â–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –º–æ–¥–µ–ª–∏

–ü—Ä–∏–º–µ–Ω–∏–º –¥–ª—è –º–æ–¥–µ–ª–µ–π –ø–∞—Ä–∞–º–µ—Ç—Ä class_weight='balanced'.

#### –ú–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –¥–µ—Ä–µ–≤–æ–º —Ä–µ—à–µ–Ω–∏–π

–û–ø—Ä–µ–¥–µ–ª–∏–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞ `max_depth` –ø–µ—Ä–µ–±–æ—Ä–æ–º –≤ —Ü–∏–∫–ª–µ –∏ –æ—Ü–µ–Ω–∫–µ –ø–æ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –Ω–∞ F1-–º–µ—Ä–µ.

```python
best_model_tree_balanced = None
best_score_f1_tree_balanced = 0

for depth in tqdm(range(1, 11)):
    model = DecisionTreeClassifier(class_weight='balanced', random_state=RANDOM, max_depth=depth)
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)
    score = f1_score(target_valid, predicted_valid)
    if score > best_score_f1_tree_balanced:
        best_model_tree_balanced = model
        best_score_f1_tree_balanced = score

print(f'–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞: {best_model_tree_balanced}\n'
      f'F1-–º–µ—Ä–∞: {best_score_f1_tree_balanced}\n'
      f'AUC-ROC: {roc_auc_score(target_valid, best_model_tree_balanced.predict_proba(features_valid)[:,1])}')
```

#### –ú–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

–û–ø—Ä–µ–¥–µ–ª–∏–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞ `n_estimators` –ø–µ—Ä–µ–±–æ—Ä–æ–º –≤ —Ü–∏–∫–ª–µ –∏ –æ—Ü–µ–Ω–∫–µ –ø–æ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –Ω–∞ F1-–º–µ—Ä–µ, –ø–∞—Ä–∞–º–µ—Ç—Ä `max_depth` –ø—Ä–∏–º–µ–º —Ä–∞–≤–Ω—ã–º 8, –∫–∞–∫ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π –¥–ª—è —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞.

```python
best_model_forest_balanced = None
best_score_f1_forest_balanced = 0

for est in tqdm(range(1, 100, 2)):
    model = RandomForestClassifier(class_weight='balanced', random_state=RANDOM, n_estimators=est, max_depth=8)
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)
    score = f1_score(target_valid, predicted_valid)
    if score > best_score_f1_forest_balanced:
        best_model_forest_balanced = model
        best_score_f1_forest_balanced = score

print(f'–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞: {best_model_forest_balanced}\n'
      f'F1-–º–µ—Ä–∞: {best_score_f1_forest_balanced}\n'
      f'AUC-ROC: {roc_auc_score(target_valid, best_model_forest_balanced.predict_proba(features_valid)[:,1])}')
```

#### –ú–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

–¢–∞–∫ –∫–∞–∫ –Ω–∞—à–∞ –æ–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞ –Ω–µ –≤–µ–ª–∏–∫–∞, –ø—Ä–∏–º–µ–Ω–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä `solver='liblinear'`.
–û–ø—Ä–µ–¥–µ–ª–∏–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ `penalty` –∏ `C` –ø–µ—Ä–µ–±–æ—Ä–æ–º –≤ —Ü–∏–∫–ª–µ –∏ –æ—Ü–µ–Ω–∫–µ –ø–æ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –Ω–∞ F1-–º–µ—Ä–µ.

```python
best_model_logistic_balanced = None
best_score_f1_logistic_balanced = 0

for penalty in ('l1', 'l2'):
    for inv_reg in tqdm(range(1, 10000, 20)):
        model = LogisticRegression(penalty=penalty, solver='liblinear', C=inv_reg/100, class_weight='balanced', random_state=RANDOM)
        model.fit(features_train, target_train)
        predicted_valid = model.predict(features_valid)
        score = f1_score(target_valid, predicted_valid)
        if score > best_score_f1_logistic_balanced:
            best_model_logistic_balanced = model
            best_score_f1_logistic_balanced = score    

print(f'–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏: {best_model_logistic_balanced}\n'
      f'F1-–º–µ—Ä–∞: {best_score_f1_logistic_balanced}\n'
      f'AUC-ROC: {roc_auc_score(target_valid, best_model_logistic_balanced.predict_proba(features_valid)[:,1])}')
```

### –£—Å—Ç—Ä–∞–Ω–µ–Ω–∏–µ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ –≤ –¥–∞–Ω–Ω—ã—Ö

–°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é `upsample` –¥–ª—è —É—Å—Ç—Ä–∞–Ω–µ–Ω–∏—è –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞.

–ù–∞ –≤—Ö–æ–¥ –æ–Ω–∞ –±—É–¥–µ—Ç –ø–æ–ª—É—á–∞—Ç—å –¥–≤–µ –≤—ã–±–æ—Ä–∫–∏: –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫, –∞ —Ç–∞–∫–∂–µ –º–Ω–æ–∂–∏—Ç–µ–ª—å –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π –¥–ª—è "–µ–¥–∏–Ω–∏—á–Ω–æ–≥–æ" –∫–ª–∞—Å—Å–∞.

–í–æ–∑–≤—Ä–∞—â–∞—Ç—å –æ–Ω–∞ –±—É–¥–µ—Ç —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –≤—ã–±–æ—Ä–∫–∏: –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫.

```python
def upsample(features, target, repeat):
    
    # –≤—ã–¥–µ–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ–∫ —Å–æ–≥–ª–∞—Å–Ω–æ –∫–ª–∞—Å—Å–∞ "0" –∏–ª–∏ "1"
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    # —É–≤–µ–ª–∏—á–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ –≤—ã–±–æ—Ä–æ–∫ –∑–∞ —Å—á—ë—Ç "1" –∫–ª–∞—Å—Å–∞ –∏ –º–Ω–æ–∂–∏—Ç–µ–ª—è "repeat"
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    # –ø–µ—Ä–µ–º–µ—à–∏–≤–∞–Ω–∏–µ –≤—ã–±–æ—Ä–æ–∫
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled
```

–ü—Ä–∏–º–µ–Ω–∏–º —Ñ—É–Ω–∫—Ü–∏—é `upsample` –∫ –æ–±—É—á–∞—é—â–µ–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∞–º, –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç `repeat` –ø—Ä–∏–º–µ–º —Ä–∞–≤–Ω—ã–º 4, —Ç–∞–∫ –∫–∞–∫ –¥–æ–ª—è –∫–ª–∞—Å—Å–∞ "1" —Å–æ—Å—Ç—è–≤–ª—è–µ—Ç 0.2 (—Ç–µ–º —Å–∞–º—ã–º —Ä–∞–∑–º–µ—Ä—ã –∫–ª–∞—Å—Å–æ–≤ —Å—Ç–∞–Ω—É—Ç —Ä–∞–≤–Ω—ã–º–∏).

<div class='alert alert-success'> ‚úîÔ∏è–†–∞–±–æ—á–∞—è —Ñ—É–Ω–∫—Ü–∏—è, —Ö–æ—Ä–æ—à–æ, —á—Ç–æ –Ω–µ –∑–∞–±—ã–ª–∏ –ø–µ—Ä–µ–º–µ—à–∞—Ç—å –¥–∞–Ω–Ω—ã–µ.
</div>

```python
# –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–Ω–∏–µ –æ–±—É—á–∞—é—â–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
features_train_balanced, target_train_balanced = upsample(features_train, target_train, 4)

# –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–Ω–∏–µ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ - –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –ø–æ–∫–∞–∑–∞–ª–æ —Ö—É–¥—à–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ, –∑–∞–∫–æ–º–µ–Ω—Ç–∏—Ä–æ–≤–∞–ª –∫–æ–¥
#features_valid, target_valid = upsample(features_valid, target_valid, 4)
```

#### –ú–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

–û–ø—Ä–µ–¥–µ–ª–∏–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞ `n_estimators` –ø–µ—Ä–µ–±–æ—Ä–æ–º –≤ —Ü–∏–∫–ª–µ –∏ –æ—Ü–µ–Ω–∫–µ –ø–æ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –Ω–∞ F1-–º–µ—Ä–µ, –ø–∞—Ä–∞–º–µ—Ç—Ä `max_depth` –ø—Ä–∏–º–µ–º —Ä–∞–≤–Ω—ã–º 9, –∫–∞–∫ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π –¥–ª—è —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞.

```python
best_model_forest = None
best_score_f1_forest = 0

for est in tqdm(range(1, 100, 2)):
    model = RandomForestClassifier(random_state=RANDOM, n_estimators=est, max_depth=9)
    model.fit(features_train_balanced_stand_scaler, target_train_balanced)
    predicted_valid = model.predict(features_valid_stand_scaler)
    score = f1_score(target_valid, predicted_valid)
    if score > best_score_f1_forest:
        best_model_forest = model
        best_score_f1_forest = score

print(f'–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞: {best_model_forest}\n'
      f'F1-–º–µ—Ä–∞: {best_score_f1_forest}\n'
      f'AUC-ROC: {roc_auc_score(target_valid, best_model_forest.predict_proba(features_valid_stand_scaler)[:,1])}')
```

#### –ú–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

–¢–∞–∫ –∫–∞–∫ –Ω–∞—à–∞ –æ–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞ –Ω–µ –≤–µ–ª–∏–∫–∞, –ø—Ä–∏–º–µ–Ω–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä `solver='liblinear'`.
–û–ø—Ä–µ–¥–µ–ª–∏–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ `penalty` –∏ `C` –ø–µ—Ä–µ–±–æ—Ä–æ–º –≤ —Ü–∏–∫–ª–µ –∏ –æ—Ü–µ–Ω–∫–µ –ø–æ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –Ω–∞ F1-–º–µ—Ä–µ.

```python
best_model_logistic = None
best_score_f1_logistic = 0

for penalty in ('l1', 'l2'):
    for inv_reg in tqdm(range(1, 10000, 20)):
        model = LogisticRegression(penalty=penalty, solver='liblinear', C=inv_reg/100, random_state=RANDOM)
        model.fit(features_train_balanced_stand_scaler, target_train_balanced)
        predicted_valid = model.predict(features_valid_stand_scaler)
        score = f1_score(target_valid, predicted_valid)
        if score > best_score_f1_logistic:
            best_model_logistic = model
            best_score_f1_logistic = score    

print(f'–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏: {best_model_logistic}\n'
      f'F1-–º–µ—Ä–∞: {best_score_f1_logistic}\n'
      f'AUC-ROC: {roc_auc_score(target_valid, best_model_logistic.predict_proba(features_valid_stand_scaler)[:,1])}')
```

–ü—Ä–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–∏ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ F1-–º–µ—Ä–µ (0.625 –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ) –ø–æ–∫–∞–∑–∞–ª–∞ –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞.

## –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏

–õ—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ F1-–º–µ—Ä–µ (0.625 –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ) –ø–æ–∫–∞–∑–∞–ª–∞ –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞, –ø—Ä–æ–∏–∑–≤–µ–¥—ë–º –µ—ë –¥–æ–æ–±—É—á–µ–Ω–∏–µ –Ω–∞ –æ–±—ä–µ–¥–∏–Ω—ë–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –∏ –∏—Ç–æ–≥–æ–≤–æ–µ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ.

### –î–æ–æ–±—É—á–µ–Ω–∏–µ –ª—É—á—à–∏—Ö –º–æ–¥–µ–ª–µ–π –ø–æ –æ–±—ä–µ–¥–∏–Ω—ë–Ω–Ω–æ–π –æ–±—É—á–∞—é—â–µ–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∞–º.

```python
# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –æ–±—É—á–∞—é—â–µ–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–æ–∫
features_train_balanced_stand_scaler = np.concatenate((features_train_balanced_stand_scaler, features_valid_stand_scaler), axis=0)
target_train_balanced = pd.concat([target_train_balanced, target_valid])
# –î–æ–æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π
best_model_forest.fit(features_train_balanced_stand_scaler, target_train_balanced)
print(best_model_forest)
```

### –ò—Ç–æ–≥–æ–≤–æ–µ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏

–õ—É—á—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –Ω–∞ F1-–º–µ—Ä–µ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –ø–æ–∫–∞–∑–∞–ª–∞ –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å –Ω–∞ –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –ü—Ä–æ–≤–µ–¥—ë–º –ø—Ä–æ–≤–µ—Ä–∫—É –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ.

```python
predicted_test = best_model_forest.predict(features_test_stand_scaler)
print(f'–ú–æ–¥–µ–ª—å {best_model_forest}:\n F1-–º–µ—Ä–∞: {f1_score(target_test, predicted_test)}\n'
      f'AUC-ROC: {roc_auc_score(target_test, best_model_forest.predict_proba(features_test_stand_scaler)[:,1])}\n'
      f'–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫:\n {confusion_matrix(target_test, predicted_test)}')
fpr, tpr, thresholds = roc_curve(target_test, best_model_forest.predict_proba(features_test_stand_scaler)[:,1])
plt.plot(fpr, tpr, color="darkorange")
plt.plot([0, 1], [0, 1], color="navy", linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f'–ö—Ä–∏–≤–∞—è –æ—à–∏–±–æ–∫')
plt.show()
```
