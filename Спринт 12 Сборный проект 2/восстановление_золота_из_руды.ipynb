### Сравним распределения  размеров гранул сырья на обучающей и тестовой выборках.

```python
for series, name in [(gold_recovery_train['rougher.input.feed_size'], 'обучающая выборка'), (gold_recovery_test['rougher.input.feed_size'], 'тестовая выборка')]:
    sns.displot(series, bins=100, kde=True).set(title=f'Размер гранул {name}', xlabel='Значение размера', ylabel='Плотность распределения')
    plt.show()
```

Внешне распределения имеют незначительные различия.

Проверим статистическим тестом равенство средних для обучающей и тестовой выборок.

Такую проверку можно провести используя t-тест, но должно быть выполнено 3 условия:
1. генеральные совокупности не должны зависеть друг от друга, будем считать это условие выполнено,
2. выборочные средние должны быть нормально распределены, данное условие также выполнено согласно ЦПТ, размеры наших выборок 14149 и 5290, что достаточно для выполнения ЦПТ о распределении средних выборок,
3. размеры выборок и дисперсии рассматриваемых генеральных совокупностей должны быть равны или нужно указать их неравество в качестве аргумента. В нашем случае размеры выборок отличаются более чем в два раза, так что условие не выполнено, нужно это будет указать в качестве аргумента.

За нулевую гипотезу примем равенство значений средних, альтернативной гипотезой будет их различие. Уровень статистической значимости примем за 0.05

```python
alpha = .05 # критический уровень статистической значимости, если p-value окажется меньше него - отвегнем нулевую гипотезу

results = st.ttest_ind(
    gold_recovery_train['rougher.input.feed_size'], 
    gold_recovery_test['rougher.input.feed_size'],
    equal_var = False
)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, p-значение: ', results.pvalue)
else:
    print('Не получилось отвергнуть нулевую гипотезу, p-значение: ', results.pvalue) 
    
print('Средний размер гранулы сырья в обучающей выборке:', gold_recovery_train['rougher.input.feed_size'].mean())
print('Средний размер гранулы сырья в тестовой выборке:', gold_recovery_test['rougher.input.feed_size'].mean())
```

Статистически различия в средних значимые, посмотрим на ящик с усами.

```python
sns.boxplot(data = (gold_recovery_train['rougher.input.feed_size'], gold_recovery_test['rougher.input.feed_size']))
plt.ylim(20, 100)
plt.show()
```

По ящику с усами видно, что минимальные и максимальные значения (за исключением выбросов) примерно совпадают, основное "тело" данных имеет большую площадь пересечения.

**Несмотря на то, что средние выборок имеют статистически значимые различия, по ящику с усами можно судить, что набор данных имеет схожую природу. Попытаемся построить модель и проверим её качество по метрике sMAPE.**

## Модель

### Функция расчёта метрики качества sMAPE
Для расчёта итоговой метрики качества, cоздадим функцию smape_rougher_final, она будет возвращять сумму 25% sMAPE на этапе флотации и 75% sMAPE финальной очистки.

```python
# функция принимает два списка: target, predict. Каждый список содержит два значения: rougher и final.
# index = 0 - rogher
# index = 1 - final

def smape_rougher_final(target, predict):
    smape =[]
    for i in (0, 1):
        smape.append(100/len(target[:,i]) * np.sum(2 * np.abs(target[:,i] - predict[:,i]) /
                                                     (np.abs(target[:,i]) + np.abs(predict[:,i]))))

#         # код ревьюера
#         print(f'Таргеты{target[i].shape}, Предикты {predict[i].shape}')
    
    return 0.25*smape[0] + 0.75*smape[1]
    
```

### Создание модели с применением кросс-валидации

#### Функция кросс-валидации

Применим для разбиение обучающей выборки метод K-fold, с делением выборки на 3 части.

Так как наши данные являются временным рядом, не будем применять перемешивание.

```python
kf5 = KFold(n_splits=5)
```

Напишем функцию валидации, которая на вход будет получать модель, а возвращать среднее сначение оценки sMAPE по валидации.

```python
# функция обучения модели с применением кросс валидации, возвращает среднюю оценку smape по выборкам
def model_validation(model):
    
    smape_cross_valid = []
    
    for train_index, test_index in kf5.split(gold_recovery_train_features):
        model.fit(gold_recovery_train_features_scaled[train_index], gold_recovery_train_targets.iloc[train_index])
        target = np.array(gold_recovery_train_targets.iloc[test_index])
        predict = model.predict(gold_recovery_train_features_scaled[test_index])
        smape_cross_valid.append(smape_rougher_final(target, predict))
    
    return mean(smape_cross_valid)    
```

<font color='green'> Кроссвалидация проведена верно!</font>

#### Модель линейной регрессии c L1-регуляризацией (Lasso)

```python
best_lasso_model = None
best_score_smape = 200  # 200 заведомо выше оценки

for i_tol in range(70, 80):
    for i_alpha in tqdm(range(55, 65)):
        lasso_model = Lasso(alpha=i_alpha*0.01, tol=i_tol*0.001, random_state=RANDOM)
        score_smape = model_validation(lasso_model)
        if score_smape < best_score_smape:
            best_lasso_model = lasso_model
            best_score_smape = score_smape

print(f'\nЛучшая модель линейной регрессии L1: {best_lasso_model}\n'
      f'Среднее значение sMAPE полученное на кросс-валидации для лучшей линейной регрессии L1 : {best_score_smape}')
```

#### Модель линейной регрессии c L2-регуляризацией (Ridge)

```python
best_ridge_model = None
best_score_smape = 200  # 200 заведомо выше оценки

for i_tol in range(1, 4, 2):
    for i_alpha in tqdm(range(6750, 6850, 1)):
        ridge_model = Ridge(alpha=i_alpha*1, tol=i_tol*0.00001, random_state=RANDOM)
        score_smape = model_validation(ridge_model)
        if score_smape < best_score_smape:
            best_ridge_model = ridge_model
            best_score_smape = score_smape
        
print(f'\nЛучшая модель линейной регрессии L2: {best_ridge_model}\n'
      f'Среднее значение sMAPE полученное на кросс-валидации для лучшей линейной регрессии L2: {best_score_smape}')
```

#### Модель решающего дерева

```python
best_tree_model = None
best_score_smape = 200  # 200 заведомо выше оценки

for depth in tqdm(range(2, 7)):
    tree_model = DecisionTreeRegressor(max_depth = depth, random_state=RANDOM)
    score_smape = model_validation(tree_model)
    if score_smape < best_score_smape:
        best_model_tree = tree_model
        best_score_smape = score_smape
        
print(f'\nЛучшая модель решающего дерева: {best_model_tree}\n'
      f'Среднее значение sMAPE полученное на кросс-валидации для лучшего дерева решений : {best_score_smape}')
```

#### Модель случайного леса
